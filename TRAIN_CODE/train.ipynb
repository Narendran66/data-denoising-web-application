```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
import librosa
import librosa.display
import matplotlib.pyplot as plt
import soundfile as sf

# Step 1: Load and Preprocess Audio Data
def load_audio(file_path, sr=16000):
    audio, _ = librosa.load(file_path, sr=sr)
    return audio

def compute_spectrogram(audio, n_fft=512, hop_length=256):
    spectrogram = librosa.stft(audio, n_fft=n_fft, hop_length=hop_length)
    magnitude = np.abs(spectrogram)  # Magnitude spectrogram
    phase = np.angle(spectrogram)    # Phase spectrogram
    return magnitude, phase

def preprocess_data(audio, n_fft=512, hop_length=256):
    magnitude, phase = compute_spectrogram(audio, n_fft, hop_length)
    return magnitude, phase

def load_dataset(noisy_folder, clean_folder, sr=16000):
    noisy_files = sorted(os.listdir(noisy_folder))  # Sort to match corresponding files
    clean_files = sorted(os.listdir(clean_folder))
    
    noisy_audios = []
    clean_audios = []
    
    for noisy_file, clean_file in zip(noisy_files, clean_files):
        noisy_path = os.path.join(noisy_folder, noisy_file)
        clean_path = os.path.join(clean_folder, clean_file)
        
        noisy_audio = load_audio(noisy_path, sr)
        clean_audio = load_audio(clean_path, sr)
        
        # Ensure both have the same length
        min_length = min(len(noisy_audio), len(clean_audio))
        noisy_audio = noisy_audio[:min_length]
        clean_audio = clean_audio[:min_length]
        
        noisy_audios.append(noisy_audio)
        clean_audios.append(clean_audio)
    
    return noisy_audios, clean_audios

# Load dataset
noisy_folder = r'F:\ABDUL\ABDUL 2024\NOISE_CANCELLATION_FOR_AUDIO\new_data\Code for Deep Denoising Autoencoder\dataset\noisy'
clean_folder = r'F:\ABDUL\ABDUL 2024\NOISE_CANCELLATION_FOR_AUDIO\new_data\Code for Deep Denoising Autoencoder\dataset\clean'

noisy_audios, clean_audios = load_dataset(noisy_folder, clean_folder)

# Combine all audio files into single arrays for processing
noisy_audio = np.concatenate(noisy_audios)
clean_audio = np.concatenate(clean_audios)

# Preprocess data (compute spectrograms)
clean_magnitude, clean_phase = preprocess_data(clean_audio)
noisy_magnitude, noisy_phase = preprocess_data(noisy_audio)

# Reshape data for the model
clean_magnitude = clean_magnitude.T[..., np.newaxis]  # Add channel dimension
noisy_magnitude = noisy_magnitude.T[..., np.newaxis]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(noisy_magnitude, clean_magnitude, test_size=0.2, random_state=42)

# Step 2: Build the Enhanced CNN Autoencoder
def build_enhanced_ddae(input_shape):
    input_layer = layers.Input(shape=input_shape)
    
    # Encoder
    x = layers.Conv1D(64, kernel_size=3, activation='relu', padding='same')(input_layer)
    x = layers.MaxPooling1D(pool_size=2, padding='same')(x)
    x = layers.Conv1D(128, kernel_size=3, activation='relu', padding='same')(x)
    x = layers.MaxPooling1D(pool_size=2, padding='same')(x)
    x = layers.Conv1D(256, kernel_size=3, activation='relu', padding='same')(x)
    encoded = layers.MaxPooling1D(pool_size=2, padding='same')(x)
    
    # Decoder
    x = layers.Conv1D(256, kernel_size=3, activation='relu', padding='same')(encoded)
    x = layers.UpSampling1D(size=2)(x)
    x = layers.Conv1D(128, kernel_size=3, activation='relu', padding='same')(x)
    x = layers.UpSampling1D(size=2)(x)
    x = layers.Conv1D(64, kernel_size=3, activation='relu', padding='same')(x)
    x = layers.UpSampling1D(size=2)(x)
    
    # Output layer with cropping to match input dimensions
    x = layers.Conv1D(1, kernel_size=3, activation='linear', padding='same')(x)
    cropped_output = layers.Cropping1D(cropping=(0, x.shape[1] - input_shape[0]))(x)
    
    autoencoder = models.Model(input_layer, cropped_output)
    autoencoder.compile(optimizer='adam', loss='mse')
    
    return autoencoder


# Ensure input shape matches the data
input_shape = (X_train.shape[1], X_train.shape[2])
ddae = build_enhanced_ddae(input_shape)
ddae.summary()

# Step 3: Train the Model
history = ddae.fit(X_train, y_train, 
                   epochs=300, 
                   batch_size=32, 
                   validation_data=(X_test, y_test))

# Save the trained model
model_save_path = 'ddae_model.h5'
ddae.save(model_save_path)
print(f"Model saved to {model_save_path}")

# Plot training loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Step 4: Denoise Audio
def denoise_audio(ddae, noisy_audio, n_fft=512, hop_length=256):
    noisy_magnitude, noisy_phase = preprocess_data(noisy_audio, n_fft, hop_length)
    noisy_magnitude = noisy_magnitude.T[..., np.newaxis]
    
    # Predict denoised magnitude spectrogram
    denoised_magnitude = ddae.predict(noisy_magnitude)
    denoised_magnitude = denoised_magnitude.squeeze().T
    
    # Reconstruct audio from magnitude and phase
    denoised_spectrogram = denoised_magnitude * np.exp(1j * noisy_phase)
    denoised_audio = librosa.istft(denoised_spectrogram, hop_length=hop_length)
    
    return denoised_audio

# Step 5: Visualize Spectrograms
def plot_spectrograms(original, noisy, denoised, sr=16000, n_fft=512, hop_length=256):
    original_magnitude, _ = compute_spectrogram(original, n_fft=n_fft, hop_length=hop_length)
    noisy_magnitude, _ = compute_spectrogram(noisy, n_fft=n_fft, hop_length=hop_length)
    denoised_magnitude, _ = compute_spectrogram(denoised, n_fft=n_fft, hop_length=hop_length)
    
    plt.figure(figsize=(15, 10))
    
    # Original Spectrogram
    plt.subplot(3, 1, 1)
    librosa.display.specshow(librosa.amplitude_to_db(original_magnitude, ref=np.max), 
                             sr=sr, hop_length=hop_length, x_axis='time', y_axis='hz', cmap='viridis')
    plt.colorbar(format='%+2.0f dB')
    plt.title("Original Clean Audio Spectrogram")
    
    # Noisy Spectrogram
    plt.subplot(3, 1, 2)
    librosa.display.specshow(librosa.amplitude_to_db(noisy_magnitude, ref=np.max), 
                             sr=sr, hop_length=hop_length, x_axis='time', y_axis='hz', cmap='viridis')
    plt.colorbar(format='%+2.0f dB')
    plt.title("Noisy Audio Spectrogram")
    
    # Denoised Spectrogram
    plt.subplot(3, 1, 3)
    librosa.display.specshow(librosa.amplitude_to_db(denoised_magnitude, ref=np.max), 
                             sr=sr, hop_length=hop_length, x_axis='time', y_axis='hz', cmap='viridis')
    plt.colorbar(format='%+2.0f dB')
    plt.title("Denoised Audio Spectrogram")
    
    plt.tight_layout()
    plt.show()

# Example usage
denoised_audio = denoise_audio(ddae, noisy_audio)

# Save the denoised audio
sf.write('denoised_audio_enhanced.wav', denoised_audio, 16000)

# Plot spectrograms
plot_spectrograms(clean_audio, noisy_audio, denoised_audio)

# Evaluate the Model
test_loss = ddae.evaluate(X_test, y_test)
print(f'Test Loss (MSE): {test_loss}')
```
